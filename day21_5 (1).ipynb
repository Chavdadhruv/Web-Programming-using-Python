```python

import requests,json
# API for website to fetch data
# url,api ,output formate will be given in offline and make slicing
api_key = "3cffebfc40dabb5b5392d32bb5c9bb69"
url = f"http://api.openweathermap.org/data/2.5/forecast?lat={lat}&lon={lon}&appid={api_key}" # API key was not defined, assuming it is the same as api_key

# print(url)
# json spelling
response=requests.get(url)
five_day=response.json()
# print(five_day)
# print(json.dumps(five_day,indent=5))#for proper indent
D={"date_time":[],"min_temp":[],"max_temp":[],"wind_speed":[],"weather":[]}
for i in five_day['list']:
  D['date_time'].append(i['dt_txt'])
  D['min_temp'].append(i['main']['temp_min'])
  D['max_temp'].append(i['main']['temp_max'])
  D['wind_speed'].append(i['wind']['speed'])
  D['weather'].append(i['weather'][0]['main'])
print(D)
```

```python
import pandas as pd
df=pd.DataFrame(D)
df
```
