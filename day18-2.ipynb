```python
%%writefile server_UDP.py
import socket
host=socket.gethostname()
port=5000
udp=socket.socket(type=socket.SOCK_DGRAM)
# SOCK_DGRAM for UDP
udp.bind((host,port))
while True:
    print("waiting for message")

    data,addr=udp.recvfrom(1024)
    print("reveived",data.decode(),"from",addr)
    msg=input("enter message")
    udp.sendto(msg.encode(),addr)
udp.close()
```

```python
%%writefile Client_UDP.py
import socket
host=socket.gethostname()
port=5000
udp=socket.socket(type=socket.SOCK_DGRAM)
# SOCK_DGRAM for UDP
while True:
    data=input("enter message")
    if not data:
        break
    udp.sendto(data.encode(),(host,port))
    print("ready to receive")
    data,adder=udp.recvfrom(1024)
    if not data:
        break
    print("received",data.decode())
udp.close()
```

```python
%%writefile server.py
import socket
host=socket.gethostname()
port=9000
s=socket.socket()
s.bind((host,port))
s.listen()
c,addr=s.accept()
print("connection establised from",str(addr))
filename=c.recv(1024).decode()
with open(filename,'wb') as f:
    while True:
        data=c.recv(1024)
        if not data:
            break
        f.write(data)
print("file received successfully")
c.close()
```

```python
%%writefile client.py
import socket
host='192.168.107.130'
port=9000
s=socket.socket()
s.connect((host,port))
filename=input("enter filename to send")
s.send(filename,encode())

with open(filename,'rb') as f:
    data=f.read(1024)
    while data:
        s.send(data)
        data=f.read(1024)
print("file sent successfully")
s.close()
```

```python
import socket
def createserver():
    s=socket.socket()
    s.bind(("localhost",8000))
    s.listen()
    while True:
        (c,addr)=s.accept()
        rd=c.recv(5000).decode()
        print(rd)
        data="HTTP/1.1 200 ok\n"
        data+="content-type:text/html;charset=utf-8\n\n"
        data+="<hmtl><body><h1>Hello World</h1></body></html>\n\n"
        c.send(data.encode())
    s.close()
print("access http://localhost:8000")
createserver()
```

```python
import socket
mysocket=socket.socket()
mysocket.connect(("gtu.ac.in",80))
cmd="GET http://www.gtu.ac.in/Circular_Acedemics.aspx HTTP/1.0\n\n".encode()
mysocket.send(cmd)
while True:
    data=mysocket.recv(1024)
    if len(data)<1:
        break
    print(data.decode(),end="")
mysocket.close()
```
