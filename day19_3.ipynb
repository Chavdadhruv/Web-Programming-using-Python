```python
import requests
from bs4 import BeautifulSoup
website="http://subslikescript.com/movies"
result=requests.get(website)
content=result.text
soup=BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
box = soup.find("article",{"class":"main-article"})
print(box.prettify())
```

```python
print(box.findAll("a",href=True))
```

```python
links=[]
for i in box.findAll("a",href=True):
    links.append(i.get("href"))
print(links)
```

```python
# OR
links=[]
for i in box.findAll("a",href=True):
    links.append(i["href"])
print(links)
```

```python
import requests
from bs4 import BeautifulSoup
root="http://subslikescript.com"
try:

  for i in links:
    website=root+i
    result=requests.get(website)
    content=result.text
    soup=BeautifulSoup(content,'html.parser')
    box = soup.find("article",{"class":"main-article"})
    title=box.find("h1").get_text()
    tran=box.find("div",{"class":"full-script"}).get_text(strip=True,separator=" ")
    print(title)
    with open(title[0:25]+".txt","w",encoding="utf-8") as f:
      f.write(tran)
except Exception as e:
  print(e)
```

```python
import requests
from bs4 import BeautifulSoup
root = "http://subslikescript.com"
for i in links:
    website = root + i
    result = requests.get(website)
    content = result.text
    soup = BeautifulSoup(content, 'html.parser')

    # Check if the article tag exists before proceeding
    box = soup.find("article", {"class": "main-article"})
    if box:  # Proceed only if the article tag is found
        title = box.find("h1").get_text()
        tran = box.find("div", {"class": "full-script"}).get_text(strip=True, separator=" ")
        print(title)
        with open(title + ".txt", "w", encoding="utf-8") as f:
            f.write(tran)
    else:
        print(f"Article tag not found for {website}. Skipping...")
```
