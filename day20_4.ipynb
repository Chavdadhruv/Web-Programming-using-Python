```python
# 1. Write a program of web scrapping using BeautifulSoup to scrape the given data from the
# following link.
# https://editorial.rottentomatoes.com/guide/popular-movies/
# On the above link, you’ll find 30 Popular movies. Scrape the Movie Title and Rating of that
# particular movie and make a Dataframe of the same.
# Your head of Dataframe should look like below:
import requests
from bs4 import BeautifulSoup
import pandas as pd
url = 'https://editorial.rottentomatoes.com/guide/popular-movies/'
response = requests.get(url)
content=response.text
soup = BeautifulSoup(content, 'html.parser')
movies=soup.find_all('div',class_='row countdown-item')
```

```python
soup
```

```python
print(movies)
```

```python
print(len(movies))
```

```python
movie=[]
rating=[]
for i in movies:
  m=i.find('h2').find('a').text.strip()
  r=i.find('span',class_='tMeterScore').text
  movie.append(m)
  rating.append(r)
```

```python
df=pd.DataFrame({'Movie':movie,'Rating':rating})
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.politifact.com/factchecks/'
response = requests.get(url)
content=response.text
soup = BeautifulSoup(content, 'html.parser')
news=soup.find_all('div','m-statement__content')
news
```

```python
title=[]
dates=[]
for i in news:
  t=i.find('div').find('a').text.strip()
  d=i.find('footer',class_='m-statement__footer').text
  title.append(t)
  dates.append(d)
```

```python
dates
```

```python
dates=soup.find_all("footer",class_="m-statement__footer")
print(dates)
```

```python
date=[]
for i in dates:
  d=i.text.strip().split("•")[-1]
  date.append(d)
print(date)
```

```python
store=soup.find_all('div',class_='m-statement__meta')
print(store)
topic=[]
for i in store:
  t=i.find('a',class_='m-statement__name').text.strip()
  topic.append(t)
topic
```

```python
df=pd.DataFrame({'Title':title,'Date':date,'Topic':topic})
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
upperframe=[]
for page in range(1,5):
    url="https://www.politifact.com/factchecks"
    result=requests.get(url).text
    soup=BeautifulSoup(result)
    frame=[]
    links=soup.find_all('li',class_='o-listicle__item')
#     print(links)
    for j in links:
        statement=j.find("div",class_="m-statement__quote").text.strip()
        link="https:\\politifact.com"
        link+=j.find("div",class_="m-statement__quote").find("a")['href']
        date=j.find("div",class_="m-statement__body").find("footer").text.split("•")[-1]
        source=j.find("div",class_="m-statement__meta").find("a").text.strip()
        label=j.find("div",class_="m-statement__content").find("img",class_="c-image__original")['alt']
        frame.append((statement,link,date,source,label))
    upperframe.extend(frame)
data=pd.DataFrame(upperframe,columns=['statement','link','date','source','label'])
data.tail(15)
```
